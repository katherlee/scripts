```python
sigma_dat = np.loadtxt("/home/katherlee/Research/gf2dmft-bdmc/bdmc/TestInput/dimer_sigma.dat")
sigma = sigma_dat[:,3].reshape((2,2,-1))
sigma_e = sigma_dat[:,4].reshape((2,2,-1))
pi_dat = np.loadtxt("/home/katherlee/Research/gf2dmft-bdmc/bdmc/TestInput/dimer_pi.dat")
pi = pi_dat[:,6].reshape((2,4,4,-1))
pi_e = pi_dat[:,7].reshape((2,4,4,-1))
```

```python
def bisect(f, x1, x2):
    y1 = f(x1)
    y2 = f(x2)
    if y1 * y2 > 0:
        raise ValueError

    y = 1
    while True:
        x = (x1 + x2) / 2
        y = f(x)
        if abs(y) < 1e-10:
            return x
        if y * y1 > 0:
            x1, y1 = x, y
        elif y * y2 > 0:
            x2, y2 = x, y
        else:
            print(y, y1, y2)
            raise ValueError

fuk = lambda x: np.mean(sigma[0,0,:] * x - h.gtau_vals[:100,0,0] * h.W_tau_vals[:100,0,0])
bisect(
    fuk,
    -2,
    0
)
```
